```python
#12th Feb Assignment Exception handling-1
"""Q1. What is an Exception in python? Write the difference between Exceptions and syntax errors"""
Ans: An Exception in Python is an error that occurs during the execution of a program. Exceptions are errors that occur 
at runtime, while syntax errors occur when the code is syntactically incorrect and cannot be parsed by the interpreter. 
Exceptions are usually caused by user input or system errors, while syntax errors are caused by incorrect code.

Exceptions are handled by the program, while syntax errors cannot be handled and must be fixed by the programmer. 
Exceptions can be caught and handled, while syntax errors cannot. Exceptions can also provide additional information 
about the error, while syntax errors do not.

"""Q2. What happens when an exception is not handled? Explain with an example"""
Ans: When an exception is not handled in Python, the program will crash and an error message will be displayed. 
For example, if you try to divide a number by zero, the program will crash and display a ZeroDivisionError:

Example:

x = 5
y = 0

z = x / y

# Output: ZeroDivisionError: division by zero

"""Q3. Which Python statements are used to catch and handle exceptions? Explain with an example"""
Ans: The Python statements used to catch and handle exceptions are try, except, and finally. 
Example: 
try: 
    # code that may throw an exception 
except ExceptionType as e: 
    # code to handle the exception 
finally: 
    # code that will always be executed, regardless of whether an exception was thrown or not
    
In this example, the code in the try block is executed. If an exception is thrown, the code in the except block is executed. Finally, regardless of whether an exception was thrown or not, the code in the finally block is executed.

For example, if we have a function that divides two numbers, we can use try and except to catch any ZeroDivisionError exceptions that may be thrown: 

def divide(x, y): 
    try: 
        result = x / y 
    except ZeroDivisionError: 
        print("Division by zero is not allowed!") 
    else: 
        print("Result is", result) 
    finally: 
        print("Executing finally clause") 
        
divide(2, 0) 

This code will print out "Division by zero is not allowed!" and "Executing finally clause".

"""Q4. Explain with an example:"""
Ans: 
A) try and else : try and else are two keywords used in Python to handle exceptions. The try block is used to test a 
block of code for errors. If an error occurs, the except block is executed.

Example:

try: 
    print(x) 
except: 
    print("An exception occurred") 

In this example, the try block checks if the variable x is defined. If it is not defined, an exception will be raised 
and the except block will be executed.

Output:
An exception occurred

B) Finally: Finally is a keyword in Python that is used to define a block of code that will be executed regardless of 
whether an exception is thrown or not. It is generally used with try/except blocks.

Example:

try:
   f = open("test.txt",encoding = 'utf-8')
   # perform file operations
finally:
   f.close()

In the above example, the file is opened and then some operations are performed on it. Finally, the file is closed 
regardless of whether an exception was thrown or not.

C) Raise: Raise is a keyword in Python that is used to raise an exception. It is used to throw an exception when a 
certain condition occurs.

For example, if we want to check if the user has entered a valid age, we can use the raise keyword to throw an 
exception if the age is less than 18.

age = int(input("Enter your age: "))
if age < 18:
    raise ValueError("Age must be greater than 18")
else:
    print("You are eligible to vote")
    
In this example, if the user enters an age less than 18, a ValueError exception is raised.

"""Q5. What are Custom Exceptions in python? Why do we need Custom Exceptions? Explain with an example"""
Ans: Custom Exceptions are user-defined exceptions in Python. They are used to create custom error messages and to 
handle errors that are specific to an application. Custom Exceptions allow developers to create their own exceptions 
that can be raised and handled in the same way as built-in exceptions.

We need Custom Exceptions because they allow us to create more meaningful error messages and to handle errors that are 
specific to our application. For example, if we have a web application that requires users to enter their age, we can 
create a custom exception called “AgeError” that will be raised when a user enters an invalid age. This will allow us 
to provide a more meaningful error message to the user, such as “Please enter a valid age”.

Example:

class AgeError(Exception):
    def __init__(self, message):
        self.message = message

def check_age(age):
    if age < 18:
        raise AgeError("You must be at least 18 years old to use this service")

try:
    check_age(17)
except AgeError as e:
    print(e.message)

Output: You must be at least 18 years old to use this service.

"""Q6. Create a custom exception class. Use this class to handle an exception."""
Ans: class CustomException(Exception): 
    def __init__(self, message): 
        self.message = message 
  
try: 
    raise CustomException("This is a custom exception") 
  
except CustomException as error: 
    print(error.message)

# Output: This is a custom exception

```
